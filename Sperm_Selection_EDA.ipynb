# Exploratory Data Analysis
# Project: Data-Driven Optimization of Sperm Selection Efficiency

```python
import numpy as np
import pandas as pd
import scipy.stats as stat
```

```python
data = pd.read_csv(r"C:\Users\santh\Downloads\Processed_Sperm_Selection_Data.csv")
data
```

```python
numeric_cols = data.select_dtypes(include = np.number).columns
numeric_cols
```

```python
categorical_cols = data.select_dtypes(include = 'object').columns
categorical_cols
```

# First Moment: Central Tendency

# Mean:

```python
for col in numeric_cols:
    print(col, ":", data[col].mean())
```

# Median:

```python
for col in numeric_cols:
    print(col, ":", data[col].median())
```

# Mode:

```python
for col in numeric_cols:
    print(col, ":", data[col].mode())
```

```python
for col in categorical_cols:
    print(col, ":", data[col].mode())
```

# Minimum:

```python
for col in numeric_cols:
    print(col, min(data[col]))
```

# Maximum:

```python
for col in numeric_cols:
    print(col, max(data[col]))
```

# Second Moment: Dispersion

# Variance:

```python
for col in numeric_cols:
    print(col, ":", data[col].var())
```

# Standard Deviation:

```python
for col in numeric_cols:
    print(col, ":", data[col].std())
```

# Range:

```python
for col in numeric_cols:
    range = max(data[col]) - min(data[col])
    print(col, ":", range)
```

```python
from scipy.stats import skew, kurtosis
```

# Third Moment: Skewness

# Skewness:

```python
from scipy.stats import skew, kurtosis
```

```python
for col in numeric_cols:
    skewness = skew(data[col])
    print(col, ":", skewness)
```

# Fourth Moment: Kurtosis

# Kurtosis

```python
for col in numeric_cols:
    print(col, ":", data[col].kurt())
```

# Summary table for four moments:


```python
Summary_table = pd.DataFrame({
    'Mean' : data[numeric_cols].mean(),
    'Median' : data[numeric_cols].median(),
    'Standard_Deviation' : data[numeric_cols].std(),
    'Variance' : data[numeric_cols].var(),
    'Range' : data[numeric_cols].max() - data[numeric_cols].min(),
    'Skewness' : skew(data[numeric_cols]),
    'Kurtosis' : data[numeric_cols].kurt()
})
```
